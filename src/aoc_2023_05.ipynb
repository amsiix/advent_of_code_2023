The almanac starts by listing which seeds need to be planted:
seeds 79, 14, 55, and 13.

The rest of the almanac contains a list of maps which describe
how to convert numbers from a source category into numbers in
a destination category. That is, the section that starts with
seed-to-soil map: describes how to convert a seed number (the
source) to a soil number (the destination). This lets the
gardener and his team know which soil to use with which seeds,
which water to use with which fertilizer, and so on.

Rather than list every source number and its corresponding
destination number one by one, the maps describe entire ranges
of numbers that can be converted. Each line within a map
contains three numbers: the destination range start, the source
range start, and the range length.

Consider again the example seed-to-soil map:

```
50 98 2
52 50 48
```

The first line has a destination range start of 50, a source
range start of 98, and a range length of 2. This line means
that the source range starts at 98 and contains two values:
98 and 99. The destination range is the same length, but it
starts at 50, so its two values are 50 and 51. With this
information, you know that seed number 98 corresponds to soil
number 50 and that seed number 99 corresponds to soil number 51.

The second line means that the source range starts at 50 and
contains 48 values: 50, 51, ..., 96, 97. This corresponds to
a destination range starting at 52 and also containing 48
values: 52, 53, ..., 98, 99. So, seed number 53 corresponds to
soil number 55.

Any source numbers that aren't mapped correspond to the same
destination number. So, seed number 10 corresponds to soil
number 10.

So, the entire list of seed numbers and their corresponding
soil numbers looks like this:

```
seed  soil
0     0
1     1
...   ...
48    48
49    49
50    52
51    53
...   ...
96    98
97    99
98    50
99    51
```

With this map, you can look up the soil number required for
each initial seed number:

    Seed number 79 corresponds to soil number 81.
    Seed number 14 corresponds to soil number 14.
    Seed number 55 corresponds to soil number 57.
    Seed number 13 corresponds to soil number 13.

````python
seed_to_soil_map = []
soil_to_fertilizer_map = []
fertilizer_to_water_map = []
water_to_light_map = []
light_to_temperature_map = []
temperature_to_humidity_map = []
humidity_to_location_map = []
````

````python
mode = None
for line in data:
    # if line contains seeds, then load the seeds into a list
    if line.startswith("seeds:"):
        seeds = [int(x) for x in line.split(":")[1].split()]
        continue
    elif line == "":
        mode = None
    elif "map" in line:
        mode = line
        continue
    

    if mode == None: # skip this line
        continue
    elif mode == "seed-to-soil map:":
        # load the seed-to-soil map into a list
        seed_to_soil_map.append([int(x) for x in line.split()])
    elif mode == "soil-to-fertilizer map:":
        soil_to_fertilizer_map.append([int(x) for x in line.split()])
    elif mode == "fertilizer-to-water map:":
        fertilizer_to_water_map.append([int(x) for x in line.split()])
    elif mode == "water-to-light map:":
        water_to_light_map.append([int(x) for x in line.split()])
    elif mode == "light-to-temperature map:":
        light_to_temperature_map.append([int(x) for x in line.split()])
    elif mode == "temperature-to-humidity map:":
        temperature_to_humidity_map.append([int(x) for x in line.split()])
    elif mode == "humidity-to-location map:":
        humidity_to_location_map.append([int(x) for x in line.split()])
        
    
````

````python

def navigate_map(map, targets):
    result = []
    for target in targets:
        r = None
        for destination, source, multiplicity in map:
            if target >= source and target <= source + multiplicity:
                result.append(destination + (target - source))
                r = 1
                break
        if r == None:
            result.append(target)
    
    return result

def get_location(seeds):
    return navigate_map(
        humidity_to_location_map, 
        navigate_map(
            temperature_to_humidity_map,
            navigate_map(
                light_to_temperature_map,
                navigate_map(
                    water_to_light_map, 
                    navigate_map(
                        fertilizer_to_water_map, 
                        navigate_map(
                            soil_to_fertilizer_map, 
                            navigate_map(seed_to_soil_map, seeds)
                            )
                        )
                    )
                )
            )
        )
    
print(min(get_location(seeds)))
````
